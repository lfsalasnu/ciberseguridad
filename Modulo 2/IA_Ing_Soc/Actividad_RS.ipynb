# Cifrado Cesar

En colab escribe un programa que cifre un archivo con cifrado Cesar.

Escribe un programa en Python que:
- Cree un archivo original con texto personalizado.
- A partir del txt cree un archivo cifrado en cesar.  

Opción de descifrado condicional.



```python
texto_original = """No podemos resolver nuestros problemas con el mismo razonamiento que usamos cuando los creamos."""
archivo_original = "mensaje.txt"
with open(archivo_original, "w") as f:
    f.write(texto_original)
```

```python
print(f"Archivo original '{archivo_original}' creado. Contenido:")
with open(archivo_original, "r") as f:
    print(f.read())
```

## Cifrar

```python
def cifrado_cesar(texto, desplazamiento):
    """Aplica cifrado César al texto con el desplazamiento especificado."""
    resultado = ""
    for char in texto:
        if char.isalpha():
            # Determinar si es mayúscula o minúscula
            ascii_base = ord('A') if char.isupper() else ord('a')
            # Aplicar desplazamiento y mantener dentro del alfabeto (26 letras)
            nuevo_char = chr((ord(char) - ascii_base + desplazamiento) % 26 + ascii_base)
            resultado += nuevo_char
        else:
            # Mantener caracteres no alfabéticos sin cambios
            resultado += char
    return resultado
```

```python
# Cifrar el contenido con desplazamiento 3
desplazamiento = 3
with open(archivo_original, "r") as f:
    contenido = f.read()
contenido_cifrado = cifrado_cesar(contenido, desplazamiento)

# Guardar el contenido cifrado en un nuevo archivo
archivo_cifrado = "mensaje_cifrado.txt"
with open(archivo_cifrado, "w") as f:
    f.write(contenido_cifrado)

print(f"\nArchivo cifrado '{archivo_cifrado}' creado. Contenido:")
with open(archivo_cifrado, "r") as f:
    print(f.read())
```

## Descifrar

```python
# Descifrar para verificar
contenido_descifrado = cifrado_cesar(contenido_cifrado, -desplazamiento)
archivo_descifrado = "mensaje_descifrado.txt"
with open(archivo_descifrado, "w") as f:
    f.write(contenido_descifrado)

print(f"\nArchivo descifrado '{archivo_descifrado}' creado. Contenido:")
with open(archivo_descifrado, "r") as f:
    print(f.read())
```
